```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Grafico de linea

x = [1, 2, 3, 4, 5]
y = [2, 3, 5 , 7 , 9]

plt.plot(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de linea')

plt.show()
```

```python
# Grafico de dispersion

x = [1, 2, 3, 4, 5]
y = [2, 3, 5 , 7 , 9]

plt.scatter(x,y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafico de dispersion')
plt.show()
```

```python
# Histograma

import numpy as np

datos = np.random.rand(1000)

plt.hist(datos, bins=20, color='skyblue', edgecolor='black')

plt.title('Histograma')

plt.show()
```

```python
# Grafico de barras

categoria = ['A', 'B', 'C']
valor = [150, 230, 180]

plt.bar(categoria, valor)
plt.title('Grafico de baras')
plt.show()
```
